```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ejemplo grafico de dispersion

data = sns.load_dataset('iris')
data
```

```python
# grafico de dispersion
sns.scatterplot(x = data['sepal_length'], y = data['sepal_width'], hue = data['species'], data=data)
plt.style.use('bmh')
plt.title('Grafico de dispersion')
plt.xlabel('Longitud del sepalo')
plt.ylabel('Anchura del sepalo')
plt.show()
```

```python
# Grafica de Ridgeplot
setosa = data[data['species'] == 'setosa']
versicolor = data[data['species'] == 'versicolor']
virginica = data[data['species'] == 'virginica']

# Configurar la figura y los ejes
fig, ax = plt.subplots(figsize=(8,6))
plt.style.use('bmh')
plt.title('Longitud del sepalo por especie')

# crear el ridgeplot
sns.kdeplot(data=setosa['petal_length'], label='setosa', ax= ax, fill=True)
sns.kdeplot(data=versicolor['petal_length'], label='versicolor', ax= ax, fill=True)
sns.kdeplot(data=virginica['petal_length'], label='virginica', ax= ax, fill=True)

# Ajuste de las leyendas
ax.legend(loc = 'upper right')
plt.show()
```

```python
# grafico de de dispersion 
sns.set_style('whitegrid')
palette = sns.color_palette('husl', 3)

sns.scatterplot(x =data['sepal_length'], y = data['sepal_width'], hue = data['species'], palette=palette, data=data)

plt.xlabel('Longitud del sepalo')
plt.ylabel('Anchura del sepalo')
plt.title('Grafico Personalizado')

plt.show()
```
