```python
# imputaicón de valores nulos
data = {
    'Nombre': ['Juan', 'Ana', 'Jose', 'Arturo'],
    'Edad': [25, 18, 23, np.nan],
    'Pais': ['MX', None, 'BR', 'MX'],
    'Ciudad': ['CDMX', 'BOG', 'RIO', None ]
}
df = pd.DataFrame(data)
df
```

```python
# rellenar valores nulos
df_fill = df.fillna(
    {
    'Edad': df['Edad'].mean(),
    'Ciudad': 'Desconocido'
}
)
df_fill
```

```python
# Eliminar filas con valores nulos
df_sin_nulos = df.dropna()
df_sin_nulos
```

```python
# Remplazar valores especificos
df_reemplazo = df.replace(
    {
        'Pais': {
            None: 'Desconocido',    
            'MX': 'Mexico',
            'CO': 'Colombia',
            'BR': 'Brasil'
        }
    }
)
df_reemplazo
```

```python
# interpolación de valores
df_interpolacion = df.copy()
df_interpolacion['Edad'] = df_interpolacion['Edad'].interpolate()
df_interpolacion
```

```python
data ={
    'Nombre':['Juan', 'Ana', 'Jose', 'Arturo', 'Pedro', 'Maria'],
    'Ciudad':['CDMX', 'CDMX', 'BOG', 'RIO', 'CDMX', 'BOG'],
    'Edad':[25, 18, 23, 27, 25, 18],
    'Puntuacion':[80, 75, 85, 90, 80, 75]
}
df = pd.DataFrame(data)
df
```

```python
# Agrupar datos 
grouped = df.groupby('Ciudad')
grouped.groups
```

```python
# Calcular la suma de las edades y puntuaciones por ciudad
aggregated_data = grouped.agg(
    {
        'Edad': 'mean',
        'Puntuacion': 'sum'
    }
)
aggregated_data
```

```python
# Definir una función para agregar datos personalizados
def rango(series):
    return series.max() - series.min()

aggregated_data_custom = grouped.agg({
    'Edad': rango,
    'Puntuacion': rango
})

aggregated_data_custom
```

```python
# Agregacion de datos en multiples columnas 
data['Categoria'] = ['A', 'B', 'A', 'B', 'A', 'B']
df = pd.DataFrame(data)
df
```

```python
# Agrupar por ciudad y categoria
grouped_multi = df.groupby(['Ciudad', 'Categoria'])
grouped_multi.groups
```

```python
# Calcular la suma de las edades y puntuaciones por ciudad y categoria
aggregated_data_multi = grouped_multi.agg(
    {
        'Edad': 'sum',
        'Puntuacion': 'sum'
    }
)
aggregated_data_multi
```
