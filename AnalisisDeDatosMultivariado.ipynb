```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
sns.set_style("darkgrid")

df = sns.load_dataset("tips")
df.head()
```

```python
# Analisi Bi-Variado 
# Muestra la relacion entre dos variables
# Relacion entre Propinas y Total de la cuenta
# Se puede ver que a medida que aumenta el total de la cuenta, tambien aumenta la propina
sns .scatterplot(x="total_bill", y="tip", data=df)
plt.show()
```

```python
# Muestras la relacion entre dos variables categoricas
# Relacion entre Propinas y Sexo
# Quien da mas propina los hombres o las mujeres
sns.barplot(x="sex", y="tip", data=df, hue="sex")
plt.show()
```

```python
# Relacion entre Total de la cuenta y Sexo
# Quien gastan mas hombres o mujeres
sns.boxplot(x="sex", y="total_bill", data=df, hue="sex")
plt.show()
```

```python
# Analisis Multi-Variado
# Muestra la relacion entre mas de dos variables
# Se comparan todas las variables que sean numericas
# En la diagonal se muestra la distribucion de cada variable
sns.pairplot(df, hue="sex")
plt.show()
```
