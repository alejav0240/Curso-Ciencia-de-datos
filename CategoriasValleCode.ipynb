```python
import os
os.makedirs("./data", exist_ok=True)
```

```python
# Definir las categorías y sus slugs
categorias = [
    ("Inteligencia Artificial y Ciencia de Datos", "inteligencia-artificial"),
    ("Desarrollo de Software y Aplicaciones", "desarrollo-software"),
    ("Ciberseguridad y Criptografía", "ciberseguridad"),
    ("Computación en la Nube y Virtualización", "computacion-nube"),
    ("Redes y Comunicaciones", "redes-comunicaciones"),
    ("Bases de Datos y Sistemas de Gestión de Información", "bases-datos"),
    ("Ingeniería de Software y Metodologías Ágiles", "ingenieria-software"),
    ("Sistemas Embebidos y Robótica", "sistemas-embebidos"),
    ("Innovación y Tecnología Emergente", "innovacion-tecnologia"),
    ("Proyectos de Impacto y Aplicaciones en la Industria", "proyectos-industria")
]
```

```python
# Crear el contenido del archivo CSV
csv_content = "\n".join([f"{nombre};{slug};0" for nombre, slug in categorias])
```

```python
# Guardar el archivo CSV
file_path = "./data/categorias_forminator.csv"
with open(file_path, "w", encoding="utf-8") as file:
    file.write(csv_content)

file_path
```
