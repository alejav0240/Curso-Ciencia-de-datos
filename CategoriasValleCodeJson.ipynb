```python
import json
import csv
import os
```

```python
def json_to_csv(json_file_path, csv_file_path):
    # Leer el archivo JSON
    with open(json_file_path, "r", encoding="utf-8") as json_file:
        data = json.load(json_file)

    # Verificar que el JSON sea una lista
    if not isinstance(data, list):
        raise ValueError("El archivo JSON debe contener una lista de objetos.")

    # Crear la carpeta del archivo CSV si no existe
    os.makedirs(os.path.dirname(csv_file_path), exist_ok=True)

    # Escribir el archivo CSV
    with open(csv_file_path, "w", encoding="utf-8", newline="") as csv_file:
        writer = csv.DictWriter(csv_file, fieldnames=data[0].keys())
        writer.writeheader()
        writer.writerows(data)

    print(f"Archivo CSV guardado en: {csv_file_path}")
```

```python
# Ejemplo de uso
json_file_path = "./data/nombres.json"  # Ruta del archivo JSON
csv_file_path = "./data/nombres.csv"   # Ruta del archivo CSV
```

```python
# Llamar a la funci√≥n
json_to_csv(json_file_path, csv_file_path)
```
