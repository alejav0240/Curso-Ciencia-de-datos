```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
sns.set_style("whitegrid")

df = sns.load_dataset("tips")
df.head()
```

```python
# Grafico de barras de propinas promedio por sexo
sns.barplot(data=df, x="sex", y="tip", hue="sex", errorbar=None, legend=False)
plt.title("Propinas promedio por sexo")
plt.show()
# Los hombres dan propinas en promedio mayores que las mujeres
```

```python
# Promedio de propinas por dia
sns.barplot(data=df, x="day", y="tip", errorbar=None,hue="day")
plt.title("Promedio de propinas por dia")
plt.colorbar
plt.show()
# Los viernes y sabados son los dias que mas propinas se dan
```

```python
sns.barplot(data=df, x="time", y="tip", errorbar=None ,hue="day")
plt.title("Promedio de propinas por dia")
plt.show()
# En la noche se dan mas propinas que en el almuerzo
```

```python
# propinas promedio por fumador 
sns.barplot(data=df, x="smoker", y="tip", errorbar=None, hue="day")
plt.title("Propinas promedio por fumador")
plt.show()
```

```python
# Grafico de Barras de propinas promedio por dia y sexo
sns.barplot(data=df, x="day", y="tip", errorbar=None, hue="sex")
plt.title("Propinas promedio por dia")
plt.show()
```

```python
# Propinas promedio por dia y sexo
sns.barplot(data=df, x="time", y="tip", errorbar=None, hue="sex")
plt.title("Propinas promedio por dia y sexo")
plt.show()
```

```python
# Grafico de Barras de Propinas promedio por dia y fumador
sns.barplot(data=df, x="smoker", y="tip", errorbar=None, hue="sex")
plt.title("Propinas promedio por dia y fumador")
plt.show()
```

```python
# Grafico de dispersion de propinas por tamanio de mesa y propinas
sns.scatterplot(data=df, x="size", y="tip", hue="size")
plt.title("Propinas por tamanio de mesa")
plt.show()
```

```python
# Grafico de barras de propinas promedio por tamanio de mesa
sns.barplot(data=df, x="size", y="tip", errorbar=None, hue="size")
plt.title("Propinas promedio por tamanio de mesa")
plt.show()
```

```python
# Calcula el porcentaje de propina y lo agrega al dataframe
df["tip_pct"] = df["tip"] / df["total_bill"] * 100
df
```

```python
# Grafico de barras de propinas por porcentaje de propina
sns.barplot(data=df, x="sex", y="tip_pct", errorbar=None, hue="day")
plt.title("Propinas por porcentaje de propina")
plt.show()
```

```python
# Grafico de barras de propinas por porcentaje de propina por tiempo
sns.barplot(data=df, x="time", y="tip_pct", errorbar=None, hue="time")
plt.title("Propinas por porcentaje de propina por tiempo")
plt.show()
```

```python
# Grafico de barras de propinas por porcentaje de propina por fumador
sns.barplot(data=df, x="smoker", y="tip_pct", errorbar=None, hue="smoker")
plt.title("Propinas por porcentaje de propina por fumador")
plt.show()
```
