```python
import matplotlib.pyplot as plt
```

```python
# Ejemplo de una linea simple 
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

plt.plot(x, y)
plt.scatter(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Grafico simple')
plt.show()
```

```python
# Grafico de barras
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [10, 20, 35, 41, 50]

plt.bar(categorias, valores)
plt.xlabel('Categoria')
plt.ylabel('Valor')
plt.title('Grafico de barras')
plt.show()
```

```python
# Uso de estilos
plt.style.use('ggplot')
plt.style.available
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

plt.plot(x, y, color='red', linestyle='dashed', linewidth=2, marker='o', markerfacecolor='blue', markersize=5)
plt.scatter(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Grafico simple')
plt.show()
```
