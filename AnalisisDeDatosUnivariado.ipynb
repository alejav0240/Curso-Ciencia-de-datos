```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
sns.set_style("darkgrid")

df = sns.load_dataset("tips")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Analisis uni-variable
# Se enfoca en una sola variable mediante la visualización de la distribución de los datos
sns.histplot(df, kde=True, x="total_bill")
plt.show()
```

```python
# Diagrama de caja 
# El primer cuartil (Q1) es el valor que deja el 25% de los datos por debajo de él.
# El segundo cuartil (Q2) es la mediana, que deja el 50% de los datos por debajo de él.
# El tercer cuartil (Q3) es el valor que deja el 75% de los datos por debajo de él.
sns.boxplot(df, x="total_bill")
plt.show()
```

```python
sns.kdeplot(df , x="total_bill")
```
