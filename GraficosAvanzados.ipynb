```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generar datos de ejemplo
np.random.seed(42)
edad_auto = np.random.randint(0, 20, size=200)
precio_auto = 30 - edad_auto + np.random.normal(-3, 3, size=200)
data = pd.DataFrame({'edad': edad_auto, 'precio': precio_auto})
data
```

```python
# crear el grÃ¡fico
fig, ax = plt.subplots(2,2,figsize=(12,10))
plt.style.use('bmh')
plt.grid(True)

# GrÃ¡fico de dispersiÃ³n
ax[0,0].scatter(data['edad'], data['precio'], color='blue', alpha=0.5)
ax[0,0].set_title('Edad vs Precio')
ax[0,0].set_xlabel('Edad del auto (aÃ±os)')
ax[0,0].set_ylabel('Precio del auto (miles de dolares)')

# Histograma de la edad
sns.histplot(data['edad'], kde=True, color='green', ax=ax[0,1])
ax[0,1].set_title('Histograma de la edad')
ax[0,1].set_xlabel('Edada del auto (aÃ±os)')
ax[0,1].set_ylabel('Frecuencia')

# Histograma del precio
sns.histplot(data['precio'], kde=True, color='blue', ax=ax[1,0])
ax[1,0].set_title('Histograma del precio')
ax[1,0].set_xlabel('Precio del auto (miles de dolares)')
ax[1,0].set_ylabel('Frecuencia')

# Eliminamos el cuadro de la esquina inferior derecha
ax[1,1].axis('off')

# Ajustamos el espacio entre los grÃ¡ficos
plt.subplots_adjust(wspace=0.3,hspace=0.3)
```
